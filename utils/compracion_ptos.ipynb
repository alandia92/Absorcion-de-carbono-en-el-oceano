```python
import seaborn as sns
import pandas as pd
from matplotlib import pyplot as plt
# Cargar los datos en un DataFrame
data = pd.read_csv('../data/ptos_predictions.csv')
df = pd.read_csv('../data/co2_predictions.csv')
```

```python
data
```

```python

fig, ax = plt.subplots(figsize=(20,10))
# Grafica los valores reales y predichos como puntos en el diagrama de dispersiÃ³n
ax.scatter(data["longitude"], data["latitude"], color="#10db9e", label="Valores reales")
ax.scatter(data["longitude_pred"], data["latitude_pred"], color="#cb6ce6", label="predicciÃ³n")

ax.set_xlabel("Longitud")
ax.set_ylabel("Latitud")
ax.legend()

plt.savefig("comparacion_valores.png", dpi=300)
plt.show()
```

```python
df
```

```python

fig, ax = plt.subplots()


ax.hist(df["tco2_real"], color="#10db9e", alpha=0.5, label="tco2_real")

ax.hist(df["tco2_pred"], color="#cb6ce6", alpha=0.5, label="tco2_pred")

ax.set_xlabel("Valor de tCO2")
ax.set_ylabel("Frecuencia")
ax.set_title("ComparaciÃ³n de valores")

ax.legend()
plt.show()

```

```python
import pandas as pd

df_melted = pd.melt(df, value_vars=["tco2_real", "tco2_pred"], var_name="Tipo", value_name="Valor")

g = sns.catplot(x="Tipo", y="Valor", kind="strip", data=df_melted, palette=["#10db9e", "#cb6ce6"])

g.set_axis_labels("", "CO2")
g.fig.suptitle("ComparaciÃ³n de valores ")

plt.show()
```

```python
import numpy as np
fig, ax = plt.subplots(figsize=(20,10))
colors = np.where(df["tco2_real"] > df["tco2_pred"], "#10db9e", "#cb6ce6")

plt.scatter(df["tco2_real"], df["tco2_pred"], color=colors)

plt.xlabel("tco2_real")
plt.ylabel("tco2_pred")
plt.title('ComparaciÃ³n entre valores')

plt.savefig("co2_comparacion_valores.png", dpi=300)
plt.show()
plt.show()
```
